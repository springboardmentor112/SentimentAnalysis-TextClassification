```python
import nltk
```

```python
import pandas as pd
```

```python
import re
```

```python
data_file = pd.read_csv('D:/Sentiment Analysis/reviews.csv')
```

**#Lower Case**

```python
data_file['content']=data_file['content'].apply(lambda x: " ".join(x.lower() for x in x.split()))
```

```python
print(data_file)
```

**#Remove Links**

```python
url_pattern = re.compile(r'https://\S+')
```

```python
def remove_urls(text):
    return url_pattern.sub('',text)
```

```python
data_file['content']=data_file['content'].apply(remove_urls)
```

```python
print(data_file)
```

**#Remove Next Lines**

```python
data_file['content']=data_file['content'].replace('\n',' ',regex=True)
```

```python
print(data_file)
```

**#Words Containing Numbers**

```python
def remove_num(text):
    return re.sub('\S*\d\S*','',text).strip()
```

```python
data_file['content']=data_file['content'].apply(remove_num)
```

```python
print(data_file)
```

**#Extra Spaces**

```python
def extra_spaces(text):
    return re.sub(' +',' ',text)
```

```python
data_file['content']=data_file['content'].apply(extra_spaces)
```

```python
print(data_file)
```

**#Special Characters**

```python
data_file['content']=data_file['content'].replace(r'[^A-Za-z0-9]+',' ',regex=True)
```

```python
print(data_file)
```

**#Removal of Stop Words**

```python
from nltk.corpus import stopwords
```

```python
stop_words=set(stopwords.words('english'))
```

```python
def remove_stop(x):
    return ' '.join([word for word in str(x).split() if word not in stop_words])
```

```python
data_file['content']=data_file['content'].apply(lambda x: remove_stop(x))
```

```python
print(data_file)
```

**#Stemming**

```python
from nltk.stem.porter import PorterStemmer
```

```python
stemmer=PorterStemmer()
```

```python
def stem_words(text):
    word_tokens=text.split()
    stems= ' '.join([stemmer.stem(word) for word in word_tokens])
    return stems
```

```python
data_file['content']=data_file['content'].apply(stem_words)
```

```python
print(data_file)
```

**#Lemmatization**

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
def lemmatize_word(text):
    word_tokens =text.split()
    lemmas= ' '.join([lemmatizer.lemmatize(word,pos='v') for word in word_tokens])
    return lemmas
```

```python
data_file['content']=data_file['content'].apply(lemmatize_word)
```

```python
print(data_file)
```
