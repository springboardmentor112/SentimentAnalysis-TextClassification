# Removing links

```python
# Load the dataset

data = pd.read_csv('reviews.csv')
```

```python
# Fill na values in the content column with an empty string

data['content'].fillna('', inplace=True)
```

```python
# Function to remove links using regex

def remove_links(text):
    # Regex pattern to match links
    pattern = r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+'
    # Replace links with an empty string
    return re.sub(pattern, '', text)
```

```python
# Applying the remove_links function to the content column using a for loop

for i in range(len(data)):
    data.at[i, 'content'] = remove_links(data.at[i, 'content'])

# Display the updated dataframe

print(data)
```

# Removing stopwords

```python
# Download NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
```

```python
# Get the set of English stopwords
stop_words = set(stopwords.words('english'))
```

```python
STOPWORDS = set(stopwords.words('english'))
def remove_stopwords(text):
    """custom function to remove the stopwords"""
    return " ".join([word for word in str(text).split() if word not in STOPWORDS])

data["content"] = data["content"].apply(lambda text: remove_stopwords(text))
print(data)
```

# Stemming

```python
from nltk.stem.porter import PorterStemmer
```

```python
data = pd.DataFrame({'content': ['']})

stemmer = PorterStemmer()

def stem_words(text):
    return " ".join([stemmer.stem(word) for word in text.split()])

data["content"] = data["content"].apply(lambda text: stem_words(text))
print(data.head())
```

# Lemmatization

```python
lemmatizer = WordNetLemmatizer()

def lemmatize_words(text):
    words = nltk.word_tokenize(text)
    lemmatized_words = [lemmatizer.lemmatize(word) for word in words]
    return ' '.join(lemmatized_words)

data['content'] = data['content'].apply(lemmatize_words)
print(data.head())
```
