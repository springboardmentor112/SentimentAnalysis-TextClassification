```python
def remove_punctuations(text):
    # Ensure the text is a string
    if isinstance(text, str):
        punctuations = string.punctuation
        translator = str.maketrans('', '', punctuations)
        return text.translate(translator)
    else:
        # Return text as is if it's not a string
        return text
```

```python
df['clean_text'] = df['content'].apply(lambda x: remove_punctuations(x))
df.head()
```

```python
stop_words = set(stopwords.words('english'))
def remove_stopwords(text):
    words = text.split()
    filtered_text = [word for word in words if word.lower() not in stop_words]
    return ' '.join(filtered_text)
```

```python
def remove_spl_char(text):
   
    if isinstance(text, str):
        text = re.sub('[A-Za-z0-9]+', ' ', text)  
        text = re.sub('\s+', ' ', text)           
        return text.strip()                       
    else:
        return text  
```

```python
df['clean_text'] = df['clean_text'].apply(lambda x: remove_spl_char(x))
```

```python
from nltk.stem.porter import PorterStemmer

# Create an instance of the PorterStemmer
ps = PorterStemmer()

def stem_words(text):
    # Stem each word in the text and join them into a single string
    return " ".join([ps.stem(word) for word in text.split()])

# Example usage
example_text = "I am running while easily managing multiple activities"
stemmed_text = stem_words(example_text)
print(stemmed_text)
```
